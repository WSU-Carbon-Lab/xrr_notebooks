```python
import matplotlib.pyplot as plt

l_edge_0 = np.arange(1000, 1015, 1)
l_edge = np.arange(1015, 1040, 0.25)
l_edge_1 = np.arange(1040, 1200, 1)
l_edge = np.concatenate((l_edge_0, l_edge, l_edge_1)).round(2)
# np.savetxt("l_edge.dat", l_edge, fmt="%f")
n = pt.xsf.index_of_refraction("C32H16N8Zn1", energy=l_edge * 1e-3, density=1.6)
carbon = pt.xsf.index_of_refraction("C", energy=l_edge * 1e-3, density=1.6)

delta_n_carbon = np.abs(n - carbon) ** 2

plt.plot(l_edge, delta_n_carbon, label="Contrast with carbon")
```
