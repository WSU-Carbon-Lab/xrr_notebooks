# Data Processing

```python
dec_beamtime = Path("C:/Users/hduva/Washington State University (email.wsu.edu)/Carbon Lab Research Group - Documents/Synchrotron Logistics and Data/ALS - Berkeley/Data/BL1101/2023Dec/XRR/Processed/ZnPc")

scans = list(dec_beamtime.glob("*/"))

for scan in scans:
    en = list(scan.glob("*/"))
    for e in en:
        pol = list(e.glob("*/"))
        for p in pol:
            try:
                data = ru.Refl(p)
                processed_dir = scan.parent
                data.to_parquet(processed_dir / f"ZnPc_{e.name}_{p.name}_refl ({scan.name}).parquet")
            except IndexError:
                continue
```

```python
xrr_1 = ru.Refl()
xrr_1.plot()
```

```python
def show_image(i):
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(xrr_1.refl.Q, xrr_1.refl.Refl, label="250 eV")
    ax.set_xlabel("Q ($\AA^{-1}$)", fontsize=16)
    ax.set_ylabel("Reflectivity", fontsize=16)
    ax.set_yscale("log")
    ax.plot(xrr_1.refl.Q.iloc[i], xrr_1.refl.Refl.iloc[i], "o")
    # Add an inset with the image of the reflectivity
    ax_inset = fig.add_axes([0.5, 0.4, 0.4, 0.4])
    ax_inset.imshow(xrr_1.images["Images"].iloc[i])
    ax_inset.axis("off")
    ax_inset.set_title("CCD Image", fontsize=12)
    # Add a second inset axis that zooms into the beamspot and the dark spot
    ax_inset2 = fig.add_axes([0.12, 0.12, 0.2, 0.2])
    ax_inset2.imshow(xrr_1.images["Beam"].iloc[i])
    ax_inset2.axis("off")
    ax_inset2.set_title("Beam", fontsize=12)


    ax_inset3 = fig.add_axes([0.3, 0.12, 0.2, 0.2])
    ax_inset3.imshow(xrr_1.images["Dark"].iloc[i])
    ax_inset3.axis("off")
    ax_inset3.set_title("Background", fontsize=12)

    fig.savefig("ZnPc_250eV_show.svg", dpi=300, bbox_inches="tight", edgecolor="white", facecolor="white")
    plt.show()

show_image(55)
```

```python
def multi_en_ploting():
    scan_number = ru.FileDialog.getDirectory()
    en = [en for en in scan_number.iterdir() if en.is_dir()]
    fig, ax = plt.subplots(
        nrows=len(en),
        figsize=(20, 4 * len(en)),
        sharex=True,
        gridspec_kw={"wspace": 0, "hspace": 0},
    )
    for e in en:
        file = scan_number / e / "100.0"
        data = ru.Refl(path=file).refl
        data.plot(
            x="Theta",
            y="Refl",
            ax=ax[en.index(e)],
            label=e.name,
            logy=True,
        )


multi_en_ploting()
```
