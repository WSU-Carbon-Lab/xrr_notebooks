```python
import pandas as pd
```

```python
df_dft = pd.read_csv("../fitting/anisotropy_weight_scan_dft.csv")
df_free = pd.read_csv("../fitting/anisotropy_weight_scan.csv")
#  Drop "Unnamed: 0" column if it exists and obj
df_dft = df_dft.loc[:, ~df_dft.columns.str.contains("^Unnamed")]
df_free = df_free.loc[:, ~df_free.columns.str.contains("^Unnamed")]
df_dft = df_dft.loc[:, ~df_dft.columns.str.contains("obj")]
df_free = df_free.loc[:, ~df_free.columns.str.contains("obj")]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the dataframes

fig, ax = plt.subplots(figsize=(8, 6))
ax2 = ax.twinx()
sns.lineplot(
    data=df_free,
    x="weight",
    y="chi2",
    label="Free Parameters",
    color="blue",
    ax=ax,
    markers=True,
    marker="o",
    markerfacecolor="white",
    markeredgecolor="black",
)
sns.lineplot(
    data=df_dft,
    x="weight",
    y="chi2",
    label="DFT Constrained",
    color="orange",
    ax=ax2,
    markers=True,
    marker="o",
    markerfacecolor="white",
    markeredgecolor="black",
)
ax.legend(loc="upper left")
ax2.legend(loc="upper right")
ax2.set_ylabel("Chi-squared (DFT Constrained)", color="orange")
ax2.tick_params(axis="y", labelcolor="orange")
ax.set_xlabel("Anisotropy Weight")
ax.set_ylabel("Chi-squared")
for ax in [ax, ax2]:
    ax.grid(True)
    ax.set_ylim(0, 2000)
```
